```python
import turtle as tt
import random
```

```python
class GreedySnake:
    def __init__(self):
        # initial velocity
        self.x_add=10
        self.y_add=0
        # initial state
        self.snake=[[0,0],[10,0],[20,0],[30,0],[40,0],[50,0]]
        self.isPause=False

        # Apple Location
        self.xapple=random.randrange(-24,24)*10
        self.yapple=random.randrange(-23,24)*10
    
    def drawSnake(self):
        for each in self.snake:
            self.square_spot(each[0],each[1],10,'black')
    
    # moving method: pop-head, append-tail
    def move_step(self):
        self.snake.pop(0)
        self.snake.append([self.snake[-1][0]+self.x_add,self.snake[-1][-1]+self.y_add])
        
    def change_direc(self,x_lab,y_lab):
        self.x_add=x_lab
        self.y_add=y_lab
        
    def is_crash(self):
        # if crash - wall
        if self.snake[-1][0]<=-250 or self.snake[-1][0]>=240 or self.snake[-1][1]<=-240 or self.snake[-1][1]>=260:
            return True
        # if crash - body
        elif self.snake[-1] in self.snake[:-1]:
            return True
        else:
            False
            
    def eat_apple(self):
        self.snake.append(self.snake[-1])
        
    def square_spot(self,x,y,width,color):
        tt.up()
        tt.goto(x,y)
        tt.down()
        tt.color(color)
        tt.begin_fill()
        # start draw
        tt.forward(width)
        tt.right(90)
        tt.forward(width)
        tt.right(90)
        tt.forward(width)
        tt.right(90)
        tt.forward(width)
        tt.right(90)
        tt.end_fill()
        tt.up()
        
    def show_apple(self):
        if self.snake[-1]==[self.xapple,self.yapple]:
            self.xapple=random.randrange(-24,24)*10
            self.yapple=random.randrange(-23,24)*10
            self.eat_apple()
        else:
            self.square_spot(self.xapple,self.yapple,10,'red')
        print('apple: %s' % [self.xapple,self.yapple])
        
    def game_loop(self):
        tt.clear()
        self.show_apple()
        self.drawSnake()
        if not self.isPause:
            self.move_step()
            print('snake: %s' % self.snake)
            print('speed: %s' % [self.x_add,self.y_add])
            print('Crash: %s' % self.is_crash())
        if not self.is_crash():
            tt.ontimer(self.game_loop,100)
        else:
            self.square_spot(self.snake[-1][0],self.snake[-1][1],10,'green')
            tt.done()
        
    def pause_resume(self):
        self.isPause = not self.isPause
        
if __name__=="__main__":
    # screen size
    tt.setup(500,500)
    tt.hideturtle()
    # don't show processes
    tt.tracer(False)
    tt.listen()
    snake=GreedySnake()
    # 若无lambda，后面对象类型是None,有lambda
    tt.onkey(lambda:snake.change_direc(-10,0),'a')
    tt.onkey(lambda:snake.change_direc(10,0),'d')
    tt.onkey(lambda:snake.change_direc(0,10),'w')
    tt.onkey(lambda:snake.change_direc(0,-10),'s')
    tt.onkey(lambda:snake.pause_resume(),' ')
    # enter game loop
    snake.game_loop()
    tt.done()
```
