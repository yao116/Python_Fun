#### 10号词云 色调

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,ImageColorGenerator

text = open('Alice.txt').read()

import imageio
mk=imageio.imread('alice.jpg')
wc = WordCloud(background_color='white',mask=mk)
wc.generate(text)

image_colors = ImageColorGenerator(mk) # 提取模板图片颜色
fig,axes=plt.subplots(1,3)
#原生词云图
axes[0].imshow(wc)
#按模板颜色生成词云，采用双线性插值的方法显色
axes[1].imshow(wc.recolor(color_func=image_colors),interpolation='bilinear')
#原模板图片
axes[2].imshow(mk,cmap=plt.cm.gray)

for ax in axes:
    ax.set_axis_off()
plt.show()

wc_color=wc.recolor(color_func=image_colors)
wc_color.to_file('10号变色.png')
```

#### 12号词云 snownlp

```python
import jieba
import wordcloud
import imageio
mk = imageio.imread("map.png")

w1=wordcloud.WordCloud(width=1000,
                      height=700,
                      background_color='white',
                      font_path='msyh.ttc', # 字体路径，微软雅黑
                      mask=mk,
                      max_words=2000,
                      scale=15) #scale越大，图像密度越大越清晰

w2=wordcloud.WordCloud(width=1000,
                      height=700,
                      background_color='white',
                      font_path='msyh.ttc', # 字体路径，微软雅黑
                      mask=mk,
                      max_words=2000,
                      scale=15) #scale越大，图像密度越大越清晰

f=open('西游记.txt',encoding='utf-8')
txt=f.read()
txtlist=jieba.lcut(txt)
positivelist=[]
negativelist=[]

# 判断>0.96为积极词，<0.06为消极词
print("开始情感分析")

import snownlp
for each in txtlist:
    each_word=snownlp.SnowNLP(each)
    feeling=each_word.sentiments
    if feeling > 0.96:
        positivelist.append(each)
    elif feeling < 0.06:
        negativelist.append(each)
    else:
        pass
    
positive_string = " ".join(positivelist)
negative_string = " ".join(negativelist)

w1.generate(positive_string)
w2.generate(negative_string)

w1.to_file('output12-positive.png')
w2.to_file('output12-negative.png')
print("词云生成完成")
```

```python
import matplotlib.pyplot as plt
plt.imshow(w1)
plt.axis("off")
plt.show()

plt.imshow(w2)
plt.axis("off")
plt.show()
```

#### 13号词云 人物分色

```python
from wordcloud import WordCloud, get_single_color_func
```

```python
class SimpleGroupedColorFunc(object):
    def __init__(self,color_to_words,default_color):
        # color_to_words: dict(str -> list(str))
        # A dictionary that maps a color to the list of words
        
        # default_color: str
        # Color that will be assigned to a word that's not a member of any value from color_to_words
        self.word_to_color={word:color
                           for (color,words) in color_to_words.items()
                           for word in words}
        self.default_color = default_color
        
    def __call__(self,word,**kwargs):
        return self.word_to_color.get(word,self.default_color)

    
class GroupedColorFunc(object):
    def __init__(self,color_to_wrords,default_color):
        self.color_func_to_words=[
            (get_single_color_func(color),set(words))
            for (color,words) in color_to_words.items()]
        
        self.default_color_func = get_single_color_func(default_color)
        
    def get_color_func(self,word):
        # Returns a single_color_func associated with the word
        try:
            color_func=next(
                color_func for (color_func,words) in self.color_func_to_words 
                if word in words)
        except StopIteration:
            color_func = self.default_color_func
        
        return color_func
        
    def __call__(self,word,**kwargs):
        return self.get_color_func(word)(word,**kwargs)
```

```python
import imageio
mk = imageio.imread("map.png")

w=wordcloud.WordCloud(width=1000,
                      height=700,
                      background_color='white',
                      font_path='msyh.ttc', # 字体路径，微软雅黑
                      mask=mk,
                      max_words=2000,
                      scale=15,font_step=1,max_font_size=60) #scale越大，图像密度越大越清晰

import jieba
f=open('三国演义.txt',encoding='utf-8')
txt=f.read()
txtlist=jieba.lcut(txt)
string = " ".join(txtlist)
w.generate(string)

color_to_words={'green':['刘备','刘玄德','孔明','诸葛孔明','张飞','赵云','马超','云长'],
                'red':['曹操','司马懿','夏侯','郭嘉','曹仁','司马昭','夏侯渊','曹真'],
                'purple':['孙权','周瑜','东吴','孙策','吕蒙','黄盖','太史慈','陆逊'],
                'pink':['董卓','袁术','袁绍','吕布','刘璋','貂蝉','刘表']}

default_color='gray' #其它词语的颜色
#构建心倒颜色规则
grouped_color_func=GroupedColorFunc(color_to_words,default_color)

w.recolor(color_func=grouped_color_func)
w.to_file('13号词云人物分色.png')
```
